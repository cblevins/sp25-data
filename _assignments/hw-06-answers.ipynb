**Age Comparison**: Calculate and write print() statements that show:

- The average age of passengers from **Ireland**
- The average age of passengers from **Scotland**.
- The difference in years between these average


For using average (median):


```python
#Your Code Here

ireland_pass=passengers_df[passengers_df['native_country']=='Ireland']
scotland_pass=passengers_df[passengers_df['native_country']=='Scotland']
average_ireland=ireland_pass['age'].median()
average_scotland=scotland_pass['age'].median()

print(f"The average age of passengers from Ireland: {average_ireland}")
print(f"The average age of passengers from Scotland: {average_scotland}")

diff_average=abs(average_ireland-average_scotland) #using absolute values to get the difference - doesn't matter which is greater
print(f"The difference in average age: {diff_average}")
```

For using average (mean):


```python
#Your Code Here

ireland_pass=passengers_df[passengers_df['native_country']=='Ireland']
scotland_pass=passengers_df[passengers_df['native_country']=='Scotland']
average_ireland=ireland_pass['age'].mean()
average_scotland=scotland_pass['age'].mean()

print(f"The average age of passengers from Ireland: {average_ireland}")
print(f"The average age of passengers from Scotland: {average_scotland}")

diff_average=abs(average_ireland-average_scotland) #using absolute values to get the difference - doesn't matter which is greater
print(f"The difference in average age: {diff_average}")
```
