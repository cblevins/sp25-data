```python
utah_df.rename(columns={'birthstate': 'birthplace'})
```

Fill NaN values with "unknown" for occupation column


```python
utah_df['occupation']=utah_df['occupation'].fillna('Unknown')
```

Searching for strings and returning rows that have them.


```python
farm_df=utah_df[utah_df['occupation'].str.contains('FARM')]
```

```python
len(farm_df)
```

```python
farm_df.groupby('occupation').count().sort_values('key', ascending=False)
```

```python
utah_df['marrystatus']=utah_df['marrystatus'].str.replace('m', 'married')
utah_df['marrystatus']=utah_df['marrystatus'].str.replace('s', 'single')
utah_df['marrystatus']=utah_df['marrystatus'].str.replace('w', 'widowed')
```

```python
def birthyear_calc(age):
    return 1880-age
```

```python
utah_df['birthyear']=utah_df['age']
```

```python
utah_df['birthyear']=utah_df['birthyear'].apply(birthyear_calc)
```

```python
utah_df['birthyear']=utah_df['birthyear'].astype('int64')
```

```python
birthsbyyear=utah_df.groupby('birthyear')['key'].count()
```

```python
birthsbyyear.plot(kind='line', title='Birth Year for People Living in Utah in 1880')
```
